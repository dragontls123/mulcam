```python
# K-Means Clustering 
#   : Sebastian Raschka, 2015, Python Machine Learning (p. 312)
# --------------------------------------------------------------------
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

#    
X, y = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std=0.5, shuffle=True, random_state=0)

#   2  
plt.figure(figsize=(8, 6))
plt.scatter(X[:,0], X[:,1], marker='o', s=100, alpha=0.5)
plt.grid()
plt.show()

# K-means    3   (k = 3)
# n_init = 10 (default) : Local min   .
#     (10 ) error      .
km = KMeans(n_clusters=3, init='random', n_init=1000, max_iter=3000, tol=1e-04, random_state=0)
km = km.fit(X)
y_km = km.predict(X)

#   
plt.figure(figsize=(8, 6))
plt.scatter(X[y_km == 0, 0], X[y_km == 0, 1], s=100, c='green', marker='s', alpha=0.5, label='cluster 1')
plt.scatter(X[y_km == 1, 0], X[y_km == 1, 1], s=100, c='orange', marker='o', alpha=0.5, label='cluster 2')
plt.scatter(X[y_km == 2, 0], X[y_km == 2, 1], s=100, c='blue', marker='v', alpha=0.5, label='cluster 3')
plt.scatter(X[y_km == 3, 0], X[y_km == 3, 1], s=100, c='red', marker='v', alpha=0.5, label='cluster 4')
plt.scatter(X[y_km == 4, 0], X[y_km == 4, 1], s=100, c='black', marker='v', alpha=0.5, label='cluster 5')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s=250, marker='+', c='red', label='centroids')
plt.legend()
plt.grid()
plt.show()
km.inertia_
```
