```python
# K-Means++ Clustering 
# mnist   
# -------------------------
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import KernelPCA
import pickle

#  mnist  .
with open('dataset/mnist.pickle', 'rb') as f:
        mnist = pickle.load(f)

# input data 
inputX = mnist.data[:5000, :]
imageX = inputX.copy()

#  0 1 2 3 6 8  
#  4 5 7 9   

# Z-score nomalization.
#    .     transpose 
# ,   transpose .
sc = StandardScaler()
inputX = sc.fit_transform(inputX.T).T

# # Kernel PCA  
pca = KernelPCA(n_components = 50, kernel='rbf')
inputX = pca.fit_transform(inputX)

# K-means++    K  .
km = KMeans(n_clusters=10, init='k-means++', n_init=3, max_iter=3000, tol=1e-01, random_state=0)
km = km.fit(inputX)
clust = km.predict(inputX)

# cluster   .
f = plt.figure(figsize=(8, 2))
for k in np.unique(clust):
    # cluster i imageX image 10 .
    idx = np.where(clust == k)[0][:10]
    
    f = plt.figure(figsize=(8, 2))
    for i in range(10):
        image = imageX[idx[i]].reshape(28,28)
        ax = f.add_subplot(1, 10, i + 1)
        ax.imshow(image, cmap=plt.cm.bone)
        ax.grid(False)
        ax.set_title(k)
        ax.xaxis.set_ticks([])
        ax.yaxis.set_ticks([])
        plt.tight_layout()

```
