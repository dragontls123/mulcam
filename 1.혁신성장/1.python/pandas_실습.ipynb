## 'And' or 'Or'
### 문제1. 얼마나 많은 사람들이 American Express 카드로 95달러 초과로 결제하는 지 알아보기

```python
con1=ecom['CC Provider']=='American Express' 
con2= ecom['Purchase Price']>95
ecom[(ecom['CC Provider']=='American Express') & (ecom['Purchase Price']>95)].count()
#ecom[con1&con2].count()
```

### 문제2. 얼마나 많은 사람들이 95달러 초과로 결제하거나 10달러 미만으로 구매하는 지 알아보기

```python
ecom[(ecom['Purchase Price']<10)|(ecom['Purchase Price']>95)].count()
```

## .apply() & Lambda
얼마나 많은 신용카드가 2025년에 만료되는 지 알아보기

## 문제1. 얼마나 많은 신용카드가 2025년에 만료되는 지 알아보기

```python
#ecom[ecom['CC Exp Date'][:2]=='25']
ecom['CC Exp Date'].apply(lambda x:x[-2:]=='25').value_counts()
```

## Lambda & .split
### 문제1. 가장 인기있는 이메일 제공자/호스트(e.g. gmail.com, yahoo.com, etc...) 상위 5 찾아내기

```python
ecom['Email'].apply(lambda x:x.split('@')[-1]).value_counts().head(5)
```
